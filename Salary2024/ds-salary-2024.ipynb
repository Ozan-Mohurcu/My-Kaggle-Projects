```python
colors = ['#2c4d8a', '#f5f5f6', '#bebebe', '#415a8a']

# Remote year'larƒ± hesaplayalƒ±m
remote_year = df.groupby(['work_year', 'work_models']).size()

# 2020, 2021, 2022, 2023, 2024 yƒ±lƒ±na ait remote √ßalƒ±≈üma oranlarƒ±nƒ± hesaplayalƒ±m
ratio_2020 = np.round(remote_year[2020].values / remote_year[2020].values.sum(), 2)
ratio_2021 = np.round(remote_year[2021].values / remote_year[2021].values.sum(), 2)
ratio_2022 = np.round(remote_year[2022].values / remote_year[2022].values.sum(), 2)
ratio_2023 = np.round(remote_year[2023].values / remote_year[2023].values.sum(), 2)
ratio_2024 = np.round(remote_year[2024].values / remote_year[2024].values.sum(), 2)

# Polar grafiƒüi olu≈üturuyoruz
fig = go.Figure()

categories = ['No Remote Work', 'Partially Remote', 'Fully Remote']

# Her yƒ±l i√ßin trace ekliyoruz ve renk paletini uyguluyoruz
fig.add_trace(go.Scatterpolar(
    r = ratio_2020, 
    theta = categories, 
    fill = 'toself', 
    name = '2020 Remote Ratio',
    line=dict(color=colors[0])  # 2020 i√ßin ilk rengi kullanƒ±yoruz
))

fig.add_trace(go.Scatterpolar(
    r = ratio_2021, 
    theta = categories,
    fill = 'toself', 
    name = '2021 Remote Ratio',
    line=dict(color=colors[1])  # 2021 i√ßin ikinci rengi kullanƒ±yoruz
))

fig.add_trace(go.Scatterpolar(
    r = ratio_2022, 
    theta = categories,
    fill = 'toself', 
    name = '2022 Remote Ratio',
    line=dict(color=colors[2])  # 2022 i√ßin √º√ß√ºnc√º rengi kullanƒ±yoruz
))

fig.add_trace(go.Scatterpolar(
    r = ratio_2023, 
    theta = categories,
    fill = 'toself', 
    name = '2023 Remote Ratio',
    line=dict(color=colors[3])  # 2023 i√ßin d√∂rd√ºnc√º rengi kullanƒ±yoruz
))

fig.add_trace(go.Scatterpolar(
    r = ratio_2024, 
    theta = categories,
    fill = 'toself', 
    name = '2024 Remote Ratio',
    line=dict(color=colors[0])  # 2024 i√ßin aynƒ± renk
))

fig.update_layout(
    title='Remote Work Ratio by Year',
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1]
        )
    ),
    showlegend=True
)

fig.show(renderer='iframe_connected')
```

```python
def Freq_df(word_list):
    Freq_dist_nltk = nltk.FreqDist(word_list)
    df_freq = pd.DataFrame.from_dict(Freq_dist_nltk, orient = 'index')
    df_freq.columns = ['Frequency']
    df_freq.index.name = 'Term'
    df_freq = df_freq.sort_values(by = ['Frequency'], ascending = False)
    df_freq = df_freq.reset_index()
    return df_freq
```

```python
def Word_Cloud(data, title):
    plt.figure(figsize = (20,15))
    wc = WordCloud(width = 1200, height = 600, max_words = 50,
               background_color = 'white',
               max_font_size = 100, random_state = 42)
    wc.generate_from_frequencies(data)
    
    plt.imshow(wc)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
freq_df = Freq_df(df['job_title'].values.tolist())
data = dict(zip(freq_df['Term'].tolist(), freq_df['Frequency'].tolist()))
data = freq_df.set_index('Term').to_dict()['Frequency']

Word_Cloud(data , 'WordCloud of job designations')
```
